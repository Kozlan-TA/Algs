```python
import matplotlib.pyplot as plt
import numpy as np

import json
import copy
```

## Задача 1 (Сортировка пузырьком по возрастанию)

```python
# Генерация
x = np.random.random(10)
y = np.random.random(10)

with open ("sample1.json", "w") as file:
    json.dump(dict(zip(x, y)), file, indent=4)

# Загрузка из файла
with open ("sample1.json", "r") as file:
    data = json.load(file)

# Сортировка
x = list(map(float, data.keys()))
y = list(map(float, data.values()))

test_x = sorted(copy.deepcopy(x))
test_y = sorted(copy.deepcopy(y))

for j in range(len(x) - 1):
    for i in range(len(x) - 1 - j):
        if x[i] > x[i + 1]:
            a = x[i]
            x[i] = x[i + 1]
            x[i + 1] = a

for j in range(len(y) - 1):
    for i in range(len(y) - 1 - j):
        if y[i] > y[i + 1]:
            a = y[i]
            y[i] = y[i + 1]
            y[i + 1] = a

assert(x == test_x)
assert(y == test_y)

print(x)
print(y)

# График
plt.plot(x, y)
plt.show()
```

## Задача 2 (Сортировка пузырьком по убыванию)

```python
# Генерация
x = np.random.random(10)
y = np.random.random(10)

with open ("sample2.json", "w") as file:
    json.dump(dict(zip(x, y)), file, indent=4)

# Загрузка из файла
with open ("sample2.json", "r") as file:
    data = json.load(file)

# Сортировка
x = list(map(float, data.keys()))
y = list(map(float, data.values()))

test_x = sorted(copy.deepcopy(x), reverse=True)
test_y = sorted(copy.deepcopy(y), reverse=True)

for j in range(len(x) - 1):
    for i in range(len(x) - 1 - j):
        if x[i] < x[i + 1]:
            a = x[i]
            x[i] = x[i + 1]
            x[i + 1] = a

for j in range(len(y) - 1):
    for i in range(len(y) - 1 - j):
        if y[i] < y[i + 1]:
            a = y[i]
            y[i] = y[i + 1]
            y[i + 1] = a

assert(x == test_x)
assert(y == test_y)

print(x)
print(y)

# График
plt.plot(x, y)
plt.show()
```

## Задача 3 (Сортировка выбором по возрастанию)

```python
# Генерация
x = np.random.random(10)
y = np.random.random(10)

with open ("sample3.json", "w") as file:
    json.dump(dict(zip(x, y)), file, indent=4)

# Загрузка из файла
with open ("sample3.json", "r") as file:
    data = json.load(file)

# Сортировка
x = list(map(float, data.keys()))
y = list(map(float, data.values()))

test_x = sorted(copy.deepcopy(x))
test_y = sorted(copy.deepcopy(y))

for i in range(len(x)):
    index = i
    for j in range(i + 1, len(x)):
        if x[j] < x[index]:
            index = j
    x[i], x[index] = x[index], x[i]

for i in range(len(y)):
    index = i
    for j in range(i + 1, len(y)):
        if y[j] < y[index]:
            index = j
    y[i], y[index] = y[index], y[i]

assert(x == test_x)
assert(y == test_y)

print(x)
print(y)

# График
plt.plot(x, y)
plt.show()
```

````python
import numpy as np
````

````python
# Задание 1 Создать вектор (одномерный массив) размера 10, заполненый нулями
a = np.zeros((10))
print(a)
````

````python
# Задание 2 Создать вектор размер 10, заполненный единицами
a = np.ones((10))
print(a)
````

````python
# Задание 3 Создать вектор размера 10, заполненный числом 2.5
a = np.full((10), 2.5)
print(a)
````

````python
# Задание 4 Создать вектор заполненный нулями, но пятый элемент равен 1
a = np.zeros(10)
a[4] = 1
print(a)
````

````python
# Задание 5 Создать вектор со значениями от 10 до 49
a = np.arange(10, 50)
print(a)
````

````python
# Задание 6 Развернуть вектор (первый становится последним)
a = np.arange(10)
b = a[::-1]
print(b)
````

````python
# Задание 7 Создать матрицу (двумерный массив) 3x3 со значениями от 0 до 8
a = np.arange(9).reshape((3, 3))
print(a)
````

````python
# Задание 8 Найти индексы ненулевых элементов в [1,2,0,0,4,0]
a = np.array([1,2,0,0,4,0])
b = np.argwhere(a != 0)
print(b)
````

````python
# Задание 9 Создать массив 3x3x3 со случайными значениями
a = np.random.random((3, 3, 3))
print(a)
````

````python
# Задание 10 Создать массив 10x10 со случайными значениями, найти минимум и максимум
a = np.random.random((10, 10))
print(a)
print(f'max: {np.max(a)}, min: {np.min(a)}')
````

````python
# Задание 11 Создать матрицу с 0 внутри и 1 на границах
a = np.ones((10, 10))
a[1: -1, 1: -1] = 0
print(a)
````

````python
# Задание 12 Создать матрицу 8x8 и заполнить ее в шахматном порядке
a = np.ones((8, 8))
a[1::2, ::2] = 0
a[::2, 1::2] = 0
print(a)
````

````python
# Задание 13 Перемножить матрицы 5x3 и 3x2
a = np.arange(15).reshape((5, 3))
b = np.arange(6).reshape((3, 2))
c = np.dot(a, b)
# Или можно a.dot(b)
print(c)
````

````python
# Задание 14 Создать случайную матрицу 5x7 целых чисел от 1 до 100 и удалить [3,5] колонки
a = np.random.random((5, 7))
print(a)
b = np.delete(a, [3, 5], axis = 1)
print(b)
````

````python
# Задание 15 Есть генератор, сделать с его помощью массив
def generate():
    for x in range(10):
            yield x
a = np.array(list(generate()))
print(a)
````

````python
# Задание 16 Создать двумерный список и преобразовать его в NumPy массив (с типом int32)
a = list([[1, 2, 3], [4, 5, 6], [7, 8, 9]])
b = np.array(a, dtype=np.int32)
print(b)
````

````python
# Задание 17 Проверить, одинаковы ли 2 NumPy массива
a = np.array([1, 2, 3, 4, 5])
b = np.array([1, 2, 3, 4, 5])
c = np.array([4, 5, 6])
print(f'a == b : {np.array_equal(a, b)}')
print(f'a == c : {np.array_equal(a, c)}')
````

````python
# Задание 18 Сделать массив неизменяемым
a = np.array([1, 2, 3])
a.flags.writeable = False
try:
    a[1] = 13
    print(a)
except Exception as exc:
    print(f"Возникло исключение {exc}")

a.flags.writeable = True
a[1] = 13
print(a)
````
